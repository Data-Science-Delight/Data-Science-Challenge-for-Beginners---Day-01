```python
# Import all necessary libraries:

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Sample Data:

data = {'First_Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
        'Marks1': [88, 74, 35, 68, 23],
        'Marks2': [95, 98, 25, 84, 50],
        'Marks3': [100, 87, 75, 91, 77]}

df = pd.DataFrame(data)
df
```

```python
# Find the average marks of each student:

df['Average'] = df[['Marks1', 'Marks2', 'Marks3']].mean(axis = 1)
```

```python
df
```

```python
# Calculate CGPA:

df['CGPA'] = df['Average']/ 9.5
df['CGPA']= df['CGPA'].round(2)
```

```python
df
```

```python
# Find Student with Maximum Marks:
df['Total_Marks'] = df[['Marks1', 'Marks2', 'Marks3']].sum(axis = 1)
df
```

```python
max_score = df.loc[df['Total_Marks'].idxmax()]
```

```python
print("Student with the highest marks: ", (max_score['First_Name']), "with marks: ", (max_score['Total_Marks']) )
```
